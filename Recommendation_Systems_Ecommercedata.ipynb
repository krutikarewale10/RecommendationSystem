```python
df = pd.read_csv('uid_users.csv')
df
```

```python
# to merge both the dataframes on common feature
df_merge = df1.merge(df, on='reviewed_by')
df_merge.head(50)
```

```python
#converting the feature datatype int
df_merge['uid']=df_merge['uid'].astype(int)
```

```python
# fucntion to get total rating for each product
combine_rating = df_merge.dropna(axis = 0, subset = ['product'])

mobile_ratingCount = (combine_rating.
     groupby(by = ['product'])['rating'].
     count().
     reset_index().
     rename(columns = {'rating': 'totalRatingCount'})
     [['product', 'totalRatingCount']]
    )
mobile_ratingCount.head()
```

```python
# to merge two dataframes 
rating_with_totalRatingCount = combine_rating.merge(mobile_ratingCount, left_on = 'product', right_on = 'product', how = 'left')
rating_with_totalRatingCount.head()
```

```python
#describe total rating count
pd.set_option('display.float_format', lambda x: '%.3f' % x)
print(mobile_ratingCount['totalRatingCount'].describe())
```

```python

print(mobile_ratingCount['totalRatingCount'].quantile(np.arange(.9, 1, .01)))
```

```python
# filter total rating counts as per the threshold
popularity_threshold = 50
rating_popular_mobile = rating_with_totalRatingCount.query('totalRatingCount >= @popularity_threshold')
rating_popular_mobile.head()
```

```python
#combined = rating_popular_mobile.merge(reviewed_by, left_on = 'reviewed_by', right_on = 'reviewed_by', how = 'left')

loc_user_rating = rating_popular_mobile[rating_popular_mobile['location'].str.contains("Kasauli|Aligarh")]
#loc_user_rating=loc_user_rating.drop('Age', axis=1)
loc_user_rating.head()
```

```python
from scipy.sparse import csr_matrix
mobile_features_df=loc_user_rating.pivot_table(index='product',columns='uid',values='rating').fillna(0)
type(mobile_features_df)
#mobile_features_df=int((mobile_features_df))
print(mobile_features_df)


mobile_features_df =mobile_features_df[~mobile_features_df.isin([np.nan, np.inf, -np.inf]).any(1)]


mobile_features_df_matrix = csr_matrix(mobile_features_df.values)
mobile_features_df_matrix
```

```python

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(mobile_features_df_matrix)

```

```python
#Randomly selecting a row 
query_index = np.random.choice(mobile_features_df.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(mobile_features_df.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 3)

```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(mobile_features_df.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, mobile_features_df.index[indices.flatten()[i]], distances.flatten()[i]))
```
